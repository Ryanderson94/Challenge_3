#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
## January 4th dataset
# Calculate the arbitrage spread for the January 4th date
arbitrage_spread_early = coinbase_sliced.loc["2018-01-04"] - bitstamp_sliced.loc["2018-01-04"]

# Filter out all negative values from the dataset to obtain a list of all positive opportunities for arbitrage
arbitrage_spread_early_positive = arbitrage_spread_early[arbitrage_spread_early > 0.00]
display(arbitrage_spread_early_positive.describe())


## February 24th dataset
# Calculate the arbitrage spread for the February 24th date
arbitrage_spread_mid = coinbase_sliced.loc["2018-02-24"] - bitstamp_sliced.loc["2018-02-24"]

# Filter out all negative values from the dataset to obtain a list of all positive opportunities for arbitrage
arbitrage_spread_mid_positive = arbitrage_spread_mid[arbitrage_spread_mid > 0.00]
display(arbitrage_spread_mid_positive.describe())


## March 26th dataset
# Calculate the arbitrage spread for the March 26th date
arbitrage_spread_late = coinbase_sliced.loc["2018-03-26"] - bitstamp_sliced.loc["2018-03-26"]

# Filter out all negative values from the dataset to obtain a list of all positive opportunities for arbitrage
arbitrage_spread_late_positive = arbitrage_spread_late[arbitrage_spread_late > 0.00]
display(arbitrage_spread_late_positive.describe())
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
## January 4th dataset
# Calculate the arbitrage spread returns for the January 4th dataset
spread_return_early = arbitrage_spread_early_positive/bitstamp_sliced.loc["2018-01-04"]

# Review the spread return DataFrame
display(spread_return_early)


## February 24th dataset
# Calculate the arbitrage spread returns for the February 24th dataset
spread_return_mid = arbitrage_spread_mid_positive/bitstamp_sliced.loc["2018-02-24"]

# Review the spread return DataFrame
display(spread_return_mid)


## March 26th dataset
# Calculate the arbitrage spread returns for the March 26th dataset
spread_return_late = arbitrage_spread_late_positive/bitstamp_sliced.loc["2018-03-26"]

# Review the spread return DataFrame
display(spread_return_late)
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
## January 4th dataset
# Filter out all spread returns that do not cover the costs of transaction (1%) to obtain a list of profitable trades from January 4th
profitable_trades_early = spread_return_early[spread_return_early > .01]

# Review the first five profitable trades
display(profitable_trades_early.head())


## February 24th dataset
# Filter out all spread returns that do not cover the costs of transaction (1%) to obtain a list of profitable trades from February 24th
profitable_trades_mid = spread_return_mid[spread_return_mid > .01]

# Review the first five profitable trades
display(profitable_trades_mid.head())


## March 26th dataset
# Filter out all spread returns that do not cover the costs of transaction (1%) to obtain a list of profitable trades from March 26th
profitable_trades_late = spread_return_late[spread_return_late > .01]

# Review the first five profitable trades
display(profitable_trades_late.head())
```

#### 4. Generate the summary statistics of your spread returns that are greater than 1%. How do the average returns compare among the three dates?

```python
## January 4th dataset
# Generate summary statistics for profitable trades from January 4th
display(profitable_trades_early.describe())


## February 24th dataset
# Generate summary statistics for profitable trades from February 24th
display(profitable_trades_mid.describe())


## March 26th dataset
# Generate summary statistics for profitable trades from March 26th
display(profitable_trades_late.describe())
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
## January 4th dataset
# Calculate the potential profits per trade for all profitable trades from January 4th
profit_early = profitable_trades_early * bitstamp_sliced.loc["2018-01-04"]

# Drop any missing values from the dataset and view sample of results
profit_per_trade_early = profit_early.dropna()
display(profit_per_trade_early.head())


## February 24th dataset
# Calculate the potential profits per trade for all profitable trades from February 24th
profit_mid = profitable_trades_mid * bitstamp_sliced.loc["2018-02-24"]

# Drop any missing values from the dataset and view sample of results
profit_per_trade_mid = profit_mid.dropna()
display(profit_per_trade_mid.head())


## March 26th dataset
# Calculate the potential profits per trade for all profitable trades from March 26th
profit_late = profitable_trades_late * bitstamp_sliced.loc["2018-03-26"]

# Drop any missing values from the dataset and view sample of results
profit_per_trade_late = profit_late.dropna()
display(profit_per_trade_late.head())
```

#### 6. Generate the summary statistics, and plot the results for each of the three DataFrames.

```python
## January 4th dataset
# Generate summary statistics for January 4th profits per trade
display(profit_per_trade_early.describe())


## February 24th dataset
# Generate summary statistics for February 24th profits per trade
display(profit_per_trade_mid.describe())


## March 26th dataset
# Generate summary statistics for March 26th profits per trade
display(profit_per_trade_late.describe())
```
